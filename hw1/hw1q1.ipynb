## Question 1, Function 1:

```
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def f(x, y):
  return 1 - np.absolute(x + y) - np.absolute(y - x)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
```

```
local minimum (also global minimum): (0, 0, -17.5)
local maximum (also global maximum): (0, 0, 0)
```

## Question 1, Function 2:
