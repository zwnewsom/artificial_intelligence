## Question 1, Function 1:

```
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def f(x, y):
  return 1 - np.absolute(x + y) - np.absolute(y - x)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig, ax = plt.subplots(1)
fig.suptitle("Function 1 in 2D", fontsize="16")
ax.contour(x, y, Z)
plt.show()

print("\n")

fig = plt.figure()
fig.suptitle("Function 1 in 3D", fontsize="16")
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
```

```
local minimum (also global minimum): (0, 0, -17.5)
local maximum (also global maximum): (0, 0, 0)
```

## Question 1, Function 2:

```
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def f(x, y):
  return (-x**2 - y**2)

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig, ax = plt.subplots(1)
fig.suptitle("Function 2 in 2D", fontsize="16")
ax.contour(x, y, Z)
plt.show()

fig = plt.figure()
fig.suptitle("Function 2 in 3D", fontsize="16")
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
```

## Question 1, Function 3:

```
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def f(x, y):
  return x**2 + y**2 - 2*x*y 

x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

fig, ax = plt.subplots(1)
fig.suptitle("Function 3 in 2D", fontsize="16")
ax.contour(x, y, Z)
plt.show()

fig = plt.figure()
fig.suptitle("Function 3 in 3D", fontsize="16")
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
```

local minimum (also global minimum): (4, 4, 0) \
local maximum (also global maximum): (-4, 4, 60)
